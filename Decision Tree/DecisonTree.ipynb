```python
# Import necessary libraries
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier 
from sklearn import tree
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
import pandas as pd
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
df.head()
```

```python
df['target'] = iris.target
```

```python
df.head()
```

```python
df.tail()
```

```python
df.sample(5)
```

```python
clf = DecisionTreeClassifier(random_state=0,criterion='entropy')
clf1 = DecisionTreeClassifier(random_state=0,criterion='gini') # default                           
```

```python
clf.fit(X, y)
```

```python
clf.predict([[5.1,3.5,1.4,0.2]])
```

```python
predicted = clf.predict(X)
```

```python
predicted
```

```python
# Evaluate the model (for example, using accuracy)
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y, predicted)
print(f"Accuracy: {accuracy}")
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
class_names = iris.target_names.tolist()
plt.figure(figsize=(12, 8))
plot_tree(clf, filled=True, feature_names=iris.feature_names, class_names=class_names)
plt.show()
```
